```python
def crearRecord(column_name, anho):
    #listas para poblar
    l_departamento = [] 
    l_municipio = []
    l_tipo = []
    l_anho=  []

    count = int(row[column_name])
    if count > 0 :
        print("Creating {} {} records. {} {}". format(column_name, count,row['departamento'], row['municipio']))
        for i in range(count):
            #print(i)
            l_departamento.append(row['departamento'].strip())
            l_municipio.append(row['municipio'].strip())                
            l_tipo.append(column_name)
            l_anho.append(anho)

        dataset = { 
                'departamento' : l_departamento , 
                'municipio' : l_municipio ,
                'tipo' : l_tipo,
                'anho' : l_anho
              }


        df_frame = pd.DataFrame(dataset , columns = ['departamento', 'municipio', 'tipo', 'anho'])
        
        print('Finish creating dataframe..')
        print('Records Created : {}'.format(df_frame.count()[0]))
        
    return df_frame 
```

```python
def insertToMongoDb(df_frame):
    # Connect to MongoDB
    print('Connecting to Cluster...')
    client =  MongoClient("mongodb+srv://maternologico:maternologico@cluster0.f2giv.mongodb.net/<dbname>?retryWrites=true&w=majority")

    print('Connecting to Database maternologico...')
    db = client['maternologico']
    collection = db['nivel_atencion']
    df_frame.reset_index(inplace=True)
    data_dict = df_frame.to_dict("records")

    # Insert collection
    print('Inserting data...')
    collection.insert_many(data_dict)

    print('Data Completed!')
    
```
